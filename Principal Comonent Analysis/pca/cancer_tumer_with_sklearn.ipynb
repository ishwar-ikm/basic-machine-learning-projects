```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cancer_tumor_data_features.csv')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pc_results = pca_model.fit_transform(scaled_X)
```

```python
plt.scatter(pc_results[:,0], pc_results[:,1])
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_, index=['PC1', 'PC2'], columns=df.columns)
```

```python
sns.heatmap(df_comp)
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_varience = []

for n in range(1, 30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)

    explained_varience.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1, 30), explained_varience)
plt.xlabel('Num of Components')
plt.ylabel('Varience Explained')
```
